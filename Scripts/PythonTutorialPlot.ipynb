```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
figure = plt.figure()
grid = figure.add_subplot(1, 1, 1)

# Major ticks every 20, minor ticks every 5
major_ticks = np.arange(0, 31, 10)
minor_ticks = np.arange(0, 31, 5)

grid.set_xticks(major_ticks)
grid.set_xticks(minor_ticks, minor=True)
grid.set_yticks(major_ticks)
grid.set_yticks(minor_ticks, minor=True)

# And a corresponding grid
grid.grid(which='both')

# Or if you want different settings for the grids:
grid.grid(which='minor', alpha=0.2)
grid.grid(which='major', alpha=0.9)

plt.show()
```
